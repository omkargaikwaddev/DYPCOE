```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df[(df['math score'] > 100) | (df['math score'] < 0)]
```

```python
df['math score'] = df['math score'].astype(float)
```

```python
df.loc[(df['math score'] > 100) | (df['math score'] < 0), 'math score'] = df['math score'].mean()
```

```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1
```

```python
outliers = df[(df['math score'] < (Q1 - 1.5 * IQR)) | (df['math score'] > (Q3 + 1.5 * IQR))]
```

```python
df['math score'] = df['math score'].clip(Q1, Q3)
```

```python
import numpy as np

df['math_log'] = np.log(df['math score'])
```
