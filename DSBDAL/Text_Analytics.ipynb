```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
text = "Text analytics is the process of analyzing text data to extract useful information."
print(text)
```

```python
#Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

```python
#POS Tagging
pos = pos_tag(tokens)
print("POS Tags:", pos)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("After Stopword Removal:", filtered_tokens)
#Explanation:
#Stopwords = common words like is, the, of
#Removed because they don’t add meaning.
```

```python
ps = PorterStemmer()
stemmed = [ps.stem(word) for word in filtered_tokens]

print("Stemmed Words:", stemmed)
#Stemming
#Reduces words to the rooot form
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("Lemmatized Words:", lemmatized)
#Lemitization
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = [
    "Text analytics extracts useful information",
    "Text data analysis is important",
    "Analytics helps in extracting insights from data"
]
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents) #Apply TF-IDF
```

```python
print("Feature Names:", vectorizer.get_feature_names_out()) #Display Feature Names
```

```python
print("TF-IDF Matrix:\n", tfidf_matrix.toarray()) #Display TF-IDF Matrix
```
