```python
text = "Text analytics is the process of analyzing text data to extract useful information."
print(text)
```

```python
#Tokenization
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
print(tokens)
```

```python
#POS TAGGING which assigns grammatical Labels like NOun, Verb, adjectives
from nltk import pos_tag

pos_tags = pos_tag(tokens)
print(pos_tags)
```

```python
#STOP WORDS REMOVAL IS, THE, OFF
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens if word.lower() not in stop_words]
print(filtered_words)
```

```python
#STEMMING: REDUCES WORDS TO THERE ROOT FORM LIKE ANALYZING -> ANALYZE
from nltk.stem import PorterStemmer

ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in filtered_words]
print(stemmed_words)
```

```python
#Lemmatization: SAME AS STEMMING
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized_words)
```

```python
#Term Frequency: DETERMINES HOW MANY TIMES A WORD APPEARS IN THE DOCUMENT
from collections import Counter

tf = Counter(filtered_words)
print(tf)
```

```python
#Inverse Document Frequency: DF measures how important a word is across multiple documents. COMMON WORDS - LOW IMPORTANCE, RARE WORD - HIGH IMP
#TF-IDF: used TF-IDF which combines TF and IDF to convert text into numerical form.
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [text]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print("Feature Names:", vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```
